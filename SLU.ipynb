```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("SLU.xlsx")
df.head()
```

```python
df.columns = [
    "Year",
    "Female_Unemployment",
    "Male_Unemployment",
    "Overall_Unemployment",
    "Inflation_Rate"
]
df.head()
```

```python
plt.figure(figsize=(8,4))
plt.plot(df["Year"], df["Overall_Unemployment"], marker="o")
plt.title("Sri Lanka Overall Unemployment Trend")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.show()

```

```python
#Early 2000s (High unemployment) - Economy relatively weak - High baseline unemployment
#2005–2012 (Sharp decline) - Economic growth - Sri Lanka experienced a significant improvement in employment during this period
#2013–2018 (Stable low unemployment) - steady state
#2019–2021 (Spike) - COVID-19 impact
#2022–2024 (Partial recovery) - Gradual economic adjustment with new jobs created - 

Key Findings
Overall unemployment was highest in the early 2000s.
A strong decline was observed from 2005 to 2012.
The labour market remained stable between 2013 and 2018.
A sharp increase occurred during 2019–2021 due to economic shocks.
Recent years show signs of recovery with some fluctuations.
```

```python
plt.figure(figsize=(8,4))

plt.plot(df["Year"], df["Male_Unemployment"], marker="o", label="Male")
plt.plot(df["Year"], df["Female_Unemployment"], marker="o", label="Female")

plt.title("Sri Lanka: Male vs Female Unemployment Trend")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df["Gender_Gap"] = df["Female_Unemployment"] - df["Male_Unemployment"]
df[["Year", "Gender_Gap"]]
```

```python
Gender-wise Key Findings
Female unemployment remains consistently higher than male unemployment throughout the period.
The gender unemployment gap is persistent and does not close even during periods of economic growth.
Economic crisis periods show a sharper increase in female unemployment compared to male unemployment.
Male unemployment remains relatively stable, indicating stronger labour market resilience.
```

```python
plt.figure(figsize=(8,4))

plt.plot(df["Year"], df["Overall_Unemployment"], marker="o", label="Unemployment")
plt.plot(df["Year"], df["Inflation_Rate"], marker="o", label="Inflation")

plt.title("Sri Lanka: Inflation vs Unemployment Trend")
plt.xlabel("Year")
plt.ylabel("Rate (%)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
📌 Inflation vs Unemployment Insights

Inflation shows high volatility compared to unemployment, which changes gradually over time.

Sharp inflation spikes do not immediately translate into proportional increases in unemployment.

Unemployment appears to respond slowly to inflationary shocks, indicating a lagged effect.

Crisis periods show a general alignment between rising inflation and increasing unemployment.

The relationship is not linear, suggesting the influence of additional economic factors.
```

```python
plt.figure(figsize=(6,4))

plt.scatter(df["Inflation_Rate"], df["Overall_Unemployment"])
plt.xlabel("Inflation Rate (%)")
plt.ylabel("Overall Unemployment (%)")
plt.title("Inflation vs Unemployment (Scatter Plot)")
plt.grid(True)
plt.show()
```

```python
df[["Inflation_Rate", "Overall_Unemployment"]].corr()
```

```python
Correlation Analysis Insights
The correlation coefficient between inflation and overall unemployment is 0.06, indicating a very weak linear relationship.
This confirms that changes in inflation do not directly translate into proportional changes in unemployment.
Visual inspection of the scatter plot supports this finding, showing no clear linear pattern.
The result suggests that unemployment is influenced by broader structural and policy-related factors beyond inflation alone.
```
